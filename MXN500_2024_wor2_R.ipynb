```r
library(tidyverse)
options(repr.plot.width=6, repr.plot.height=2)
```

## Activity 1 – Reading in data

```r

Ecology <- read_csv("Ecologyv2.csv")
Ecology
```

## Activity 2 – Building a plot

```r

ggplot(data=Ecology, aes(x=DO)) +
  geom_histogram()
```

```r
ggplot(data=Ecology, aes(x=DO)) +
  geom_histogram(binwidth = 1)
```

```r
ggplot(data = Ecology, aes(x = DO))+
  geom_histogram(binwidth = 1)+
  xlab("Dissolved oxygen (mg/L)")+
  ylab("Number of observations")
```

```r
ggplot(data=Ecology, aes(x=Atyidae)) +
  geom_bar() +
  xlab("Atyidae")
```

```r
ggplot(data = Ecology, aes(x = Atyidae))+
  geom_histogram(binwidth = 20, color = "grey20", fill = "lightskyblue")+
  scale_x_continuous(breaks = seq(0,160,20))+ #This changes the numbers on the x-axis
  theme_bw()
```

```r
# ggplot(data=Ecology, aes(x=Atyidae)) +
#   geom_histogram(aes(fill=..count..))
# using `..count..` is depreciated, updated code to use `after_stats(count)` instead
ggplot(data=Ecology, aes(x=Atyidae)) +
  geom_histogram(aes(fill=after_stat(count)))
```

## Activity 3 – Bivariate plot

```r
ggplot(data=Ecology,
       aes(x=Turbidity, y= DO)) +
  xlab("Turbidity")+
  geom_point()
```

```r
ggplot(data=Ecology,
       aes(x=Turbidity, y= DO)) +
  geom_point()+
  geom_smooth()
```

```r
ggplot(data=Ecology,
       aes(x=Turbidity, y= DO)) +
  geom_point(aes(color=Location)) +
  scale_colour_brewer(palette = "Dark2")+
  theme_bw()
#Note, here it's important that we use one of the "qualitative" colour schemes as we are applying this to nominal categorical data.
```

## Activity 4 – Assessing a plot

```r
install.packages("vegan")
library(vegan)
```

```r
names(Ecology)
Ecology_species <- select(Ecology, Parastacidae:Plecoptera)
```

```r
Ecology <- mutate(Ecology, TaxonRichness = specnumber(Ecology_species))
select(Ecology, Site, TaxonRichness)
```

```r
ggplot(data = Ecology, # Use the Ecology dataframe
       aes(x = Site, #Set the x-axis to show Site
           y = TaxonRichness)) + #Set the y-axis to show TaxonRichness
  geom_col(aes(fill = DO)) + #change the colour of each column (in a column plot) to map to DO
  theme_bw() + #Change the theme to black and white
  scale_fill_gradientn(colours = rainbow(7), #change the colour scheme to be rainbow
                       name = "Dissolved Oxygen") + #change the legend label
  facet_wrap(facets = ~ Location, #Create a different plot for each Location
             nrow = 1, #Include only one row of plots
             scales="free_x") + #have a different x-axis on each plot
  ylab("Taxon Richness (species)") #change the y-axis label
```

## Activity 5 – Making a better graph

```r
ggplot(data = Ecology, aes(x = DO, y = TaxonRichness))+
  geom_point(aes(colour = Location), size = 2.5)+
  xlab("Dissolved Oxygen (mg/L)")+
  ylab("Taxon Richness")+
  scale_color_brewer(palette = "Dark2")+
  theme_bw()
```
